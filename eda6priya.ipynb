```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sns
```

```python
data=pd.read_csv('udemycourses .csv')
data
```

```python
#basic filtering data:-
# 1. Display Top 10 Rows of the DataSet
data.head(10)
```

```python
# 2. Check Last 5 rows of the Dataset
data.tail(5)
```

```python
# 3. Find Shape of our DataSet (number of rows & columns)
print("Rows:", data.shape[0])
print("Columns:", data.shape[1])
```

```python
#4. Are there any missing values?
data.isnull().sum()
```

```python
#5. What is the statistical summary of numerical columns?
data.describe()
```

```python
#6 data info 
data.info()
```

```python
#Univariate Analysis Questions
#df.columns
#7. Number of Courses Per Subject
data['subject'].value_counts()
```

```python
sns.countplot(x='subject', data=data)
pl.xticks(rotation=60)
pl.show()
```

```python
#8. Which Levels Udemy Provides Courses For
data['level'].value_counts()
```

```python
sns.countplot(x='level', data=data)
pl.xticks(rotation=60)
pl.show()
```

```python
#9. Count of Paid & Free Courses
data['is_paid'].value_counts()
```

```python
sns.countplot(x='is_paid', data=data)
pl.show()
```

```python
#10. Which Courses Have More Lectures (Free or Paid)
data.groupby('is_paid')['num_lectures'].sum()
```

```python
sns.barplot(x='is_paid', y='num_lectures', data=data)
pl.show()
```

```python
#11. Which Courses Have Higher Number of Subscribers (Free or Paid)
data.groupby('is_paid')['num_subscribers'].sum()
```

```python
sns.barplot(x='is_paid', y='num_subscribers', data=data)
pl.show()
```

```python
#12. Which Level Has Highest Number of Subscribers
data.groupby('level')['num_subscribers'].sum().sort_values(ascending=False)
```

```python
sns.barplot(x='level', y='num_subscribers', data=data)
pl.xticks(rotation=60)
pl.show()
```

```python
#13. Most Popular Course Title
data[data['num_subscribers'] == data['num_subscribers'].max()]['course_title']
```

```python
#14. Top 10 Most Popular Courses

top10 = data.sort_values(by='num_subscribers', ascending=False).head(10)

top10[['course_title', 'num_subscribers']]
```

```python
sns.barplot(x='num_subscribers', y='course_title', data=top10)
pl.show()
```

```python
#15. Courses Having Highest Number of Reviews

data[data['num_reviews'] == data['num_reviews'].max()]['course_title']
```

```python
#16. Does Price Affect Number of Reviews?

sns.scatterplot(x='price', y='num_reviews', data=data)

pl.show()
```

```python
#17. Total Number of Courses Related to Python
python_courses = data[data['course_title'].str.contains('Python', case=False)]

print("Total Python Courses:", python_courses.shape[0])
```

```python
#18. Top 10 Most Popular Python Courses
top_python = data[data['course_title'].str.contains('Python', case=False)]

top_python = top_python.sort_values(by='num_subscribers', ascending=False).head(10)

top_python[['course_title', 'num_subscribers']]
```

```python
sns.barplot(x='num_subscribers', y='course_title', data=top_python)

pl.show()
```
